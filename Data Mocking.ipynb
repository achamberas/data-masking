```python
import datetime
from datetime import datetime
from faker import Faker
fake = Faker(['en-US', 'en_US', 'en_US', 'en-US'])

def mock_data(function, params):
    if function == 'ascii_free_email':
        return fake.ascii_free_email()
    elif function == 'city':
        return fake.city()
    elif function == 'company':
        return fake.company()
    elif function == 'first_name':
        return fake.first_name()
    elif function == 'last_name':
        return fake.last_name()
    elif function == 'phone_number':
        return fake.phone_number()
    elif function == 'secondary_address':
        return fake.secondary_address()
    elif function == 'street_address':
        return fake.street_address()
    elif function == 'suffix':
        return fake.suffix()
    elif function == 'word':
        return fake.word()
    elif function == 'language_name':
        return fake.language_name()
    elif function == 'country_code':
        return fake.country_code()
    elif function == 'bothify':
        return fake.bothify(text=params)
    elif function == 'date_between_dates':
        param_list = params.split(',')
        start_date = datetime.strptime(param_list[0], '%Y-%m-%d')
        end_date = datetime.strptime(param_list[1], '%Y-%m-%d')
        date_format = param_list[2]
        return fake.date_between_dates(start_date, end_date).strftime(date_format)
    elif function == 'random_element':
        param_list = params.split(',')
        return fake.random_element(elements=param_list)
    elif function == 'constant':
        return params
    elif function == 'column':
        return None
    elif function == 'lookup':
        return None
    else:
        return None
```

```python
import pandas as pd
import random

sheet_name = "mocking_rules"
rules = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
df_rules = pd.read_csv(rules)

df = pd.read_csv('HUMANA_MEMBER.txt', sep='|')
changes = []

def change_data(c, r):
    function = df_rules[df_rules['FIELD NAME'].eq(c)]['masking function'].iloc[0]
    params = df_rules[df_rules['FIELD NAME'].eq(c)]['masking params'].iloc[0]
    prev = df[c][r]
    df[c][r] = mock_data(function, params)
    change = {"row":r,"mbr_pid":df["MBR_PID"][r],"change":c,"prev":prev,"new":df[c][r]}
    return change
    
for r in df.index:
    
    # track row, MBR_ID, change, prev and new value
    rand = random.random()
    if rand < 0.0001:
        #delete
        changes.append({"row":r,"mbr_pid":df["MBR_PID"][r],"change":"delete","prev":"NA","new":"NA"})
        df.drop(axis=0,index=r,inplace=True)
    elif rand < 0.0004:
        # change address
        c = "MBR_ADDR1"
        change = change_data(c,r)
        changes.append(change)
    elif rand < 0.0011:
        # change phone number
        c = "MBR_PHONE"
        change = change_data(c,r)
        changes.append(change)
    elif rand < 0.0013:
        # change last name
        c = "MBR_LNAME"
        change = change_data(c,r)
        changes.append(change)
    elif rand < 0.0015:
        # change email
        c = "EMAIL_ID"
        change = change_data(c,r)
        changes.append(change)
    elif rand < 0.0017:
        # change practice
        c = "CENTER_NM"
        change = change_data(c,r)
        changes.append(change)
    elif rand < 0:
        # change Do Not Call
        c = "MBR_ADDR1"
        #df[c][r] = mock_data(function, params)
        #changes.append({"row":r,"mbr_pid":df["MBR_PID"][r],"change":"do not call","prev":"NA","new":df[c][r]})

df.to_csv('HUMANA_MEMBER2.txt', sep='|', index=False)
df_changes = pd.DataFrame(changes)
df_changes.to_csv('HUMANA_MEMBER_changes.txt', sep=',', index=False)
```
